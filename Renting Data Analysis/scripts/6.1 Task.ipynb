```python
# Check for any mixed-type columns

for col in df_rent.columns.tolist():
    weird = (df_rent[[col]].applymap(type) != df_rent[[col]].iloc [0].apply(type)).any(axis = 1)
    if len(df_rent[weird]) > 0:
        print(col) 
```

```python
#changing the data type 
df_rent['Parking Space']= df_rent['Parking Space'].astype('category')
df_rent['Pets   Allowed']= df_rent['Pets   Allowed'].astype('category')
df_rent['Type of Flat']= df_rent['Type of Flat'].astype('category')
df_rent['Date of posted Ads']= df_rent['Date of posted Ads'].astype('object')
```

```python
df_rent.dtypes
```

```python
# Change column Name
df_rent.rename(columns = {'Service Charge':'Service_charge'},inplace=True)
df_rent.rename(columns = {'Base  Rent':'Base_rent'},inplace=True)
df_rent.rename(columns = {'Total Rent':'Total_rent'},inplace= True)
df_rent.rename(columns = {'Calculated Total Rent':'Calculated_total_rent'},inplace=True)
df_rent.rename(columns = {'Newly Construction':'Newly_construction'},inplace=True)
df_rent.rename(columns = {'Constructed    Year ':'Constructed_year'},inplace=True)
df_rent.rename(columns = {'Pets   Allowed':'Pets_allowed'},inplace=True)
df_rent.rename(columns = {'Living space Area':'Living_space_area'},inplace=True)
```

```python
#changing data type to reduce memory problem
df_rent['Scout ID']= df_rent['Scout ID'].astype('object')
df_rent['Number of Rooms']= df_rent['Number of Rooms'].astype('category')
df_rent['Living Space Range']= df_rent['Living Space Range'].astype('category')
```
