### Creating Correlation heatmap using Seaborn version.

```python
df.columns
```

```python
# creating a subset excluding not need rows
sub= df[['Service_charge', 'Base_rent',
       'Calculated_total_rent', 'Newly_construction',
       'Balcony', 'Price trend', 'Parking Space', 'Kitchen', 'cellar',
       'Living_space_area', 'Lift', 'Type of Flat',
       'Number of Rooms', 'Living Space Range',]]
```

```python
sub.head()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

The correlation between total calculated rent and other service are showing moderate relationship.
Also, Many newly constructed building have more facilities as compare to old constructed buildings.

### Categorial Plot

```python
sns.histplot(df['Price trend'], bins = 20, kde = True)
```

```python
df.loc[df['Price trend'] < 3, 'Price category'] = 'Low price'
df.loc[(df['Price trend'] >= 3) & (df['Price trend'] < 7), 'Price category'] = 'Moderate price'
df.loc[df['Price trend'] > 7, 'Price category'] = 'High price'
```

```python
df['Price category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style='ticks')
g = sns.catplot(x='Living_space_area', y='Price trend', hue='Price category', data=df)
```

Question to explore:

The rent not depend upon the facilities and showing weak relationship. I think the rent is based on location.

-How many rooms are mostly preferred?
People are preferring two room sets and living space area and price are correlated.

-Is newly construction flat are having more facilities than old constructed flats?
The newly construction flats are providing more facilities than old contructed and also affect price trends.
